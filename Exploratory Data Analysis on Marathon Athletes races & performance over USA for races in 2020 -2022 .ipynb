```python
#what ages performance best in  races (20 +races with mins)
df3.query('Race_length=="50km"').groupby('Athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).head(15)
```

```python
#what ages performance worst in  races (20 +races with mins)
df3.query('Race_length=="50mi"').groupby('Athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>19').head(15)
```

```python
#what ages performance worst in  races (10 +races with mins)
df3.query('Race_length=="50km"').groupby('Athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>9').head(15)
```

```python
#seasons for the data - runners slow in winter 
#spring:3-5
#summer-6-8
#autumn/fall- 9-12
#winter:12-2

#split between 2 decimals 

df3['race_month']= df3['race_day'].str.split('.').str.get(1).astype(int)

df3


```

```python
df3['Race_Season']=df3['race_month'].apply(lambda x:'Winter' if x > 11 else 'Fall' if x>8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
df3.head(30)
```

```python

df3.groupby('Race_Season')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)

#Conclusion : Performance speed drop  in summer could be due to temperatures & best seen in Spring
```
